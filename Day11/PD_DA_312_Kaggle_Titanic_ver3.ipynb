```python
import math
```

```python
age_series = df.Age.dropna().apply(lambda age: math.floor(age / 10) * 10)
age_series.name = "Age_Group"
```

```python
age_series[:10]
```

```python
df = pd.concat(
    [df, age_series],
    axis=1, # 0으로하면 행으로 합치기, 1은 열으로 합치기
)
df.head(10)
```

```python
df.Age_Group = df.Age_Group.fillna("Uncategoized") # NaN으로 되어있는 값을 Uncategoized로 변경해서 필터를 실행
```
